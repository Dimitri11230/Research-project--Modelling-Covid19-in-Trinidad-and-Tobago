```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
data=np.loadtxt("actualdraft2.txt",skiprows=1, usecols=np.arange(0,4))

actualactive =data[0:,3]
actualactive=np.array(actualactive)
print(actualactive)
actualactive.shape
```

This is where we learn parameters for the model. Fitting to all data



```
def getsusceptible(x):
  # Total population, N.
  N = 1400000
  # Initial number of infected and recovered individuals, I0 and R0.
  I0, R0 = 33, 140
  # Everyone else, S0, is susceptible to infection initially.
  S0 = N - I0 - R0
  # Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
  beta, gamma = x[0], x[1]
  # A grid of time points (in days)
  t = np.linspace(0, 93, 93)

  # The SIR model differential equations.
  def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

  # Initial conditions vector
  y0 = S0, I0, R0
  # Integrate the SIR equations over the time grid, t.
  ret = odeint(deriv, y0, t, args=(N, beta, gamma))
  S, I, R = ret.T
  I=np.array(I)
  difference = actualactive-I
  difference=np.absolute(difference)
  summ =np.sum(difference)
  return summ


from scipy import optimize 

optimize.brute(getsusceptible, ((0,2),(0.025,0.1)),Ns=100)
#optimize.brute(getsusceptible, ((0,1),(0,1)),Ns=100,finish="None")
#optimize.minimize(getsusceptible,[1.5,1.5],method="CG")
```
